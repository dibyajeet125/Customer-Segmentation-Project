```r
# customer gender visualization

ggplot(df, aes(x=Gender, fill=Gender))+
  geom_bar(col = rainbow(2))+
  xlab("Female/Male")
```

```r
df1=table(df$Gender)
df1

per=df1/sum(df1)*100
per
```

```r
pie(df1,
    labels = paste(names(df1), "\n", round(per, 1), "%"),
    col = rainbow(2),
    main = "Gender Distribution")
```

```r
# customer age visualization

n_bins = 10
ggplot(df, aes(x = Age)) +
  geom_histogram(bins = n_bins, fill = "lightblue", color = "black") +
  ggtitle("Age Distribution")
```

```r
ggplot(df, aes(y=Age))+
  geom_boxplot()
```

```r
summary(df$Age)
```

```r
# customer annual income visualisation

n_bins = 10
ggplot(df, aes(x = Annual.Income..k..)) +
  geom_histogram(bins = n_bins, fill = "lightblue", color = "black") +
  ggtitle("Annual Income Distribution")
```

```r
ggplot(df, aes(x = Annual.Income..k..)) +
  geom_density(fill = "yellow", color = "orange") +
  ggtitle("Annual Income Distribution")
```

```r
summary(df$Annual.Income..k..)
```

```r
# customer spending score visualisation

ggplot(df, aes(y=Spending.Score..1.100.))+
  geom_boxplot()
```

```r
n_bins = 10
ggplot(df, aes(x = Spending.Score..1.100.)) +
  geom_histogram(bins = n_bins, fill = "lightblue", color = "black") +
  ggtitle("Spending Score Distribution")
```

```r
summary(df$Spending.Score..1.100.)
```
